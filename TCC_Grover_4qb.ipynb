```python
# from mod_gate_colours import gate_colours
gate_colours = {
    "displaycolor": {
        "x": [ # gate name
            "#271080", # box color (red)
            "#FFFFFF" # box text color (white)
        ],
        "id": [ # gate name
            "#666666", # box color (gray)
            "#FFFFFF" # box text color (white)
        ],
        "h": [ # gate name
            "#166866", # box color (red)
            "#FFFFFF" # box text color (white)
        ],
        "mcz": [ # gate name
            "#004aad", # box color (red)
            "#ffffff" # box text color (white)
        ],
        "z": [ # gate name
            "#1f91dd", # box color (red)
            "#FFFFFF" # box text color (white)
        ],
    },
}
```

```python
def inicializa_s(qc, qubits):
    for i in qubits:
        # qc.reset(i)
        qc.h(i)
    return qc

def mcz_circ(qc):
    qc.h(3)
    qc.mcx([0, 1, 2],3)
    qc.h(3)
    return qc

def oraculo_Uw(qc, qubits, winner, QuantumCircuit):
    for i, index in enumerate(winner):
        if index == "0":
            qc.x(i)
        # else: qc.id(i)

    mcz_gate = mcz_circ(qc = QuantumCircuit(4,name="mcz")).to_instruction() # Para agrupar as portas em uma unica
    qc.append(mcz_gate,[i for i in range(4)])

    for i, index in enumerate(winner):
        if index == "0":
            qc.x(i)
        # else: qc.id(i)

def difusor_Us(qc, qubits):
    for i in qubits:
        qc.h(i)
        qc.x(i)

    mcz_gate = mcz_circ(qc = QuantumCircuit(4,name="mcz")).to_instruction() # Para agrupar as portas em uma unica
    qc.append(mcz_gate,[i for i in range(4)])

    for i in qubits:
        qc.x(i)
        qc.h(i)
    return qc
```

```python
def sort_dict_by_binary_keys(input_dict):
    return dict(sorted(input_dict.items(), key=lambda item: int(item[0], 2)))

def reverse_qubits(input_dict):
    reversed_dict = {}
    for key, value in input_dict.items():
        reversed_key = ''.join(reversed(key))
        reversed_dict[reversed_key] = value
    return reversed_dict
```

```python
def plota_hist(chaves, valores, num_shots, backend_name, i):
    backend_name = " ".join(backend_name.split("_")).title()

    fig, ax = plt.subplots(figsize=(10, 6))
    plt.bar(chaves, valores, width=.5)#, ax=ax)  # Passa o eixo (axis) diretamente para plot_histogram

    # Adicionar título e rótulos manualmente usando matplotlib
    ax.set_title(f'Histograma de {backend_name} (OptimLvl: {i})', fontsize= 16) # Para usar quando sem supressao
    # ax.set_title(f'Histograma de {backend_name} (OptimLvl: {i} + DD)', fontsize= 16) # Para usar quando tiver supressao
    ax.set_xlabel('Resultados', loc='center', fontsize= 13)
    ax.set_ylabel('Contagens', fontsize= 14)
    ax.set_xticklabels(chaves, rotation=60, ha='center')  # Adjust rotation angle and alignment
    for i, valor in enumerate(valores):
        ax.text(i, valor + 1, str(round(valor*(100/num_shots),2)) + ' %', ha='center', va='bottom', fontsize=10)

    # Mostrar o gráfico
    plt.show()
```

```python
n = 4 # numero de qubits
qubits = QuantumRegister(n, 'qubit')
bits= ClassicalRegister(n, 'bit')

N = 2**n
N_it = (round((pi/4)*sqrt(N)))

marked_state = "1111"

QC = QuantumCircuit(qubits, bits)

inicializa_s(QC, range(n))

for _ in range(N_it):
    # QC.barrier([i for i in range(n)])
    oraculo_Uw(QC, range(n), marked_state, QuantumCircuit)
    difusor_Us(QC, (range(n)))

QC.barrier([i for i in range(n)])

QC_for_statevector = QC.copy()

QC.measure([0,1,2,3],[0,1,2,3])

QC.draw(
    output='mpl',
    justify="left",
    fold=-1,
    initial_state=True,
    reverse_bits=False,
    style=gate_colours
    )
```

```python
# Criar uma sessão com o backend escolhido
with Session(backend=backend_HW) as session:
    # Inicializar o Sampler dentro da sessão]
    # options = Options()

    sampler = Sampler()
    sampler.options.dynamical_decoupling.enable = True
    #
    num_shots = 20000
    # Executar o circuito
    job_0 = sampler.run([compiled_QC_0], shots=num_shots)
    job_1 = sampler.run([compiled_QC_1], shots=num_shots)
    job_3 = sampler.run([compiled_QC_3], shots=num_shots)

    # Monitorar o progresso do job
    print(f"Job ID: {job_0.job_id()}")
    print(f"Job ID: {job_1.job_id()}")
    print(f"Job ID: {job_3.job_id()}")
    print()
    print(f"Detalhes: {session.details()}")
    print()
    while True:
        status_0 = job_0.status()
        print(f"Status 0: {job_0.status()}")
        print()

        status_1 = job_1.status()
        print(f"Status 1: {job_1.status()}")
        print()

        status_3 = job_3.status()
        print(f"Status 3: {job_3.status()}")
        print()

        # Verificar se o job foi concluído ou falhou
        if status_1 and status_0 and status_3 in ["DONE", "CANCELLED", "ERROR"]:
            break

        # Esperar alguns segundos antes de verificar novamente
        time.sleep(300)
```
